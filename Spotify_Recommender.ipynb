## Text Preprocessing

```python
import nltk
import re
nltk.download('punkt')
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()

# Data processing
def text_process(input_string):
  
  new_sent = re.sub('[^a-zA-Z0-9]', ' ',input_string) # data cleaning step

  # Lowering is Important to remove the stop words
  new_sent = new_sent.lower()
  # converting to words
  new_sent = new_sent.split()
  
  # It take only "Normal Words" no "Stop Words"
  filtered_word = [stemmer.stem(word) for word in new_sent if not word in set(nltk.corpus.stopwords.words('english'))]
  return ' '.join(filtered_word)
```

```python
df_new['text']=df_new['text'].apply(text_process)
df_new
```

## Find Similarity Between the Songs

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(max_features=5000,stop_words='english',analyzer='word')

# Here we take all the songs tags and from that we Will extract 5000 most frequent Words
# tags 1 + tags2 + tags3 + tags4 + ............(All Songs tags)
vector = tfidf.fit_transform(df_new['text']).toarray()
vector.shape
```

```python
vector
```

```python
# 5000 most frequent Words
# Numeric value may represent song's of 1930s or 1920s
tfidf.get_feature_names_out()
```

```python
len(tfidf.get_feature_names_out())
```

```python
# Here We have Total 10000 songs 
# Here we find Cosine Similarity between a Particular song and all other song
# Thats why size of the matrix becomes 10000 X 10000
similarity_matrix = cosine_similarity(vector)
similarity_matrix 
```

```python
similarity_matrix.shape
```

## Checking For Reccomendation

```python
# Enumerate the array to get both Cosine_Similarity_values and indices
enumerated_array = list(enumerate(similarity_matrix[0]))
enumerated_array
```

```python
def recommend(song_name):
    # Get an Index of the Movie
    f=0
    # Here we take an "Index" of the movie not "Id" of the movie
    
    # Getting an Index of Song Searched by User
    index=df_new[df_new['song']==song_name].index[0]
    # Now From Similarity Matrix We Extract the Similarity of that song with Other's Song.
    enumerated_array = list(enumerate(similarity_matrix[index]))
    
    # Sort the array based on the values in descending order
    sorted_distance = sorted(enumerated_array, key=lambda x: x[1], reverse=True)
    
    # Get the top five distances with their indices
    top_five_elements = sorted_distance[:6]

    # here index means movie "index" not "id"
    # here value means "Cosine_Similarity_Value", Check it in above cell
    for index,value in top_five_elements:
        if f!=0: # In order to eliminate Same Movie.
            print(df_new['song'].iloc[index])
        f=1
```
